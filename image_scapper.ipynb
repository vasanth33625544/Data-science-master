```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import logging
import os
```

```python
save_dir="images/"
```

```python
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
response=requests.get("https://www.google.com/search?q=elon+musk&safe=active&sxsrf=APwXEddwBWkyYi7p3u4D15Wly4VSzbm7WQ:1680706629786&source=lnms&tbm=isch&sa=X&ved=2ahUKEwioi4-NgJP-AhWjT2wGHR2QDYEQ_AUoAnoECAEQBA&biw=1223&bih=609&dpr=1.5")
```

```python
response
```

```python
soup=BeautifulSoup( response.content,'html.parser')
```

```python
image_tags=soup.find_all('img')
```

```python
del image_tags[0]
```

```python
len(image_tags)
```

```python
image_tag_mango=[]
for i in image_tags:
    image_url=i['src']
    image_data=requests.get(image_url).content
    mydict={"index":image_url,"image":image_data}
    image_tag_mango.append(mydict)
    with open(os.path.join(save_dir,f"elon_musk_{image_tags.index(i)}.jpg"),"wb") as f:
        f.write(image_data)
```

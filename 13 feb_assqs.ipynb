## Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.


*ArithmeticError is simply an error that occurs during numeric calculations.

*ArithmeticError types in Python include:

   * OverFlowError
   * ZeroDivisionError
   * FloatingPointError

```python
# ZeroDivisionError
try:
  arithmetic = 5/0
  print(arithmetic)
except ArithmeticError:
  print('You have just made an Arithmetic error')

#Mathematically, dividing an integer by zero is wrong, and that is the reason why Python crashes the program and returns an error message.
```

## Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

* The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

```python
# lists
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")
```

```python
pylenin_info = {'name': 'Lenin Mishra', 'age': 28,'language': 'Python'}
try:
    pylenin_info['names']
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

```python
import sys
try:
    from exception import myexception
except ImportError as e:
    print(e,{e.__class__})
```

* Common Causes of ModuleNotFoundError

* Common Causes of ImportError
    * The module has dependencies that are missing
    * The module has dependencies that cannot be imported
